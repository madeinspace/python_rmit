# Practical Exercise Accessing the value of a cell
Today, we will explore the abalone dataset. But first, you will need to retrieve data. This process is the same as what we did in Week 1.


## load csv to pandas dataframe

```python
import pandas as pd
```

```python
abalone_p = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
```

```python
abalone = pd.read_csv(
    abalone_p, 
    sep=',', 
    decimal='.', 
    header=None, 
    names=['Sex', 'Length', 'Diameter', 'Height', 
            'Whole weight', 'Shucked weight', 
            'Viscera weight', 'Shell weight', 'Rings'])
abalone
```

## Manipulating the data
There are a variety of ways to access the value of a cell in a DataFrame.

For example, let's access the 4th line of the `‘Sex’` column of the abalone data.

Now, it's your turn to fill in the code by following the instruction below.

### Access by specifying the column first, then the line of the row:

```python
abalone['Sex'][3]
```

### Access by using `.loc` attribute:

```python
abalone.loc[3, 'Sex']
```

### Access by using `.iloc` attribute

```python
abalone.iloc[3, 0]
```

### Now, let’s access a limited section of rows by using masks. Check the diameter information of those abalone whose rings are `21`.

```python
ring_mask = abalone['Rings'] == 21
abalone[ring_mask]['Diameter']
```

### Check mean value by grouping a feature, e.g. check the mean value when grouping abalone by `‘Rings’`:

```python
abalone.groupby(['Rings']).mean()
```

### Sort the abalone data by `‘Length’`:

```python
abalone.sort_values(['Length'])
```

## Solution

### Access by specifying the column first, then the line of the row:

```python
abalone['Sex'][3]
```

### Access by using `.loc` attribute:

```python
abalone.loc[3, 'Sex']
```

### Access by using `.iloc` attribute

```python
abalone.iloc[3, 0]
```

### Now, let’s access a limited section of rows by using masks. Check the diameter information of those abalone whose rings are `21`.

```python
mask_ring21 = abalone['Rings'] == 21
abalone.loc[mask_ring21,'Diameter']
```

### Check mean value by grouping a feature, e.g. check the mean value when grouping abalone by `‘Rings’`:

```python
abalone.groupby(['Rings']).mean()
```

### Sort the abalone data by `‘Length’`:

```python
abalone.sort_values(by='Length').head()
```
